Uma plataforma monitora e controla a potência elétrica entregue a veículos em um eletroposto comercial. Suponha que, em uma determinada sessão de recarga controlada, a potência instantânea total entregue pelo sistema, em kW, seja modelada por

P(t) = 18 + (24t)/(t+2), t = 0

em que t representa o tempo, em minutos, contado a partir do início da sessão.

 

Perguntas:

Elabore um gráfico da função.
Qual o valor do limite de P(t) para t tendendo a 0+ e para t tendendo ao infinito?  Explique o que esses valores representam para o comportamento da recarga ao longo do tempo.
Calcule a variação média da potência entre t = 1 e t = 3. O que esses resultados indicam para o gerenciamento da infraestrutura?
Determine a derivada P'(t). Explique, com suas palavras, o significado físico dessa derivada no contexto da sessão de recarga.
Compare P'(1) e P'(6).

Analise a partir de P'(t) em que intervalos a potência está aumentando e diminuindo. Construa o gráfico de P'(t).

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# 1. Vetor de tempo
# -------------------------
t = np.linspace(0.0001, 50, 500) # para o grafico

# -------------------------
# 2. Função P(t)
# -------------------------
P = 18 + (24*t)/(t+2)

# -------------------------
# 3. Limites
# -------------------------
lim_t0 = P[0]
lim_inf = P[-1]

print(f"Limite t → 0+ ~= {lim_t0 :.0f}")
print(f"Limite t → infinito ~= {lim_inf :.0f}")

# -------------------------
# 4. Variação média (t=1 até t=3)
# -------------------------
t1, t2 = 1, 3

P1 = 18 + (24*t1)/(t1+2)
P2 = 18 + (24*t2)/(t2+2)

variacao_media = (P2 - P1) / (t2 - t1)

print("P(1) =", P1)
print("P(3) =", P2)
print("Variação média =", variacao_media)

# -------------------------
# 5. Derivada numérica
# -------------------------
def dP(t):
    numerador = 24 * (t + 2) - 24 * t
    denominador = (t + 2)**2
    return numerador / denominador

dP1 = dP(1)
dP6 = dP(6)

dP_vals = dP(t)

print("P'(1) =", dP1)
print("P'(6) =", dP6)
# -------------------------
# 6. Gráfico de P(t)
# -------------------------
plt.figure(figsize=(10,5))
plt.plot(t, P, label="P(t)")
plt.axhline(lim_inf, linestyle='--', color='red', label='Limite (t → infinito)')
plt.title("Potência ao longo do tempo")
plt.xlabel("Tempo (min)")
plt.ylabel("Potência (kW)")
plt.legend()
plt.grid()

# -------------------------
# 7. Gráfico de P'(t)
# -------------------------
plt.figure(figsize=(10,5))
plt.plot(t, dP_vals, label="P'(t)", color='green')
plt.axhline(0, linestyle='--', color='red')
plt.title("Taxa de variação da potência")
plt.xlabel("Tempo (min)")
plt.ylabel("kW/min")
plt.legend()
plt.grid()

plt.show()
```

### 1  Explicação dos limites

Quando t → 0+, P(t) ~= 18:
→ A recarga começa com aproximadamente 18 kW
Quando t → infinito, P(t) ~= 41:
→ A potência se estabiliza em aproximadamente 41 kW

o sistema tem um limite máximo a potência cresce e depois estabiliza

### 2  Interpretação da variação média

A potência aumenta em média 3.2 kW/min entre 1 e 3 minutos. Isso indica que o sistema ainda está em fase de crescimento

### 3  Significado da derivada

P′(t) representa a taxa de variação da potência. Logo, quão rápido a potência está aumentando

### 4  Comparação P′(1) e P′(6)

No início, aumento rápido (5.333333333333333)
Depois, aumento lento (0.75)

A recarga desacelera ao longo do tempo
